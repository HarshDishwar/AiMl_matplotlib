```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
```

```python
time_data = {'facebook':120, 'youtube':90, 'instagram':40, 'snapchat':28} # i have take x,y from a dictionary where x is the key,y is the value ,
                                                                           # you can also use x=(facebook,insta,snap,watsap) and  y=(120,90,40,28)
social_media = list(time_data.keys())
time_spend = list(time_data.values())
```

```python
style.use("ggplot")
plt.figure(figsize=(5,5))
# for different color use a list . multicolor=('red','green','blue',pink) then pass this list into the color ie,
plt.bar(social_media,time_spend,width=0.5,color='grey')  # color='multicolor'
plt.xlabel('Social Media')
plt.ylabel('Time in Minutes')
plt.title('Average time spend on social media')
```

```python
fb = (120, 90, 110, 100, 95)
yt = (80, 90, 75, 85, 95)
ig = (50, 40, 55, 35, 45)
sc = (20, 30, 35, 25, 28)
```

```python
xpos=np.arange(len(fb))  #np.arange(len(fb)) generates an array of integers starting from 0 up to len(fb) - 1, which is the length of the array fb.
xpos
```

```python
style.use('ggplot')
plt.figure(figsize=(8,5))
barWidth = 0.2
plt.bar(xpos,fb,width=barWidth,color='red',label=fb)
plt.bar(xpos+0.2,yt,width=barWidth,color='green',label=yt)
plt.bar(xpos+0.4,ig,width=barWidth,color='blue',label=ig)
plt.bar(xpos+0.6,sc,width=barWidth,color='yellow',label=sc)
plt.xlabel('days')
plt.ylabel('Time spend in minutes')
plt.legend
plt.show()
```

```python
 fb = (120, 90, 110, 100, 95)
yt = (80, 90, 75, 85, 95)
ig = (50, 40, 55, 35, 45)
x = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
xpos = np.arange(len(x))
barWidth = 0.4
plt.figure(figsize=(8,5))
plt.bar(xpos, fb, color='royalblue', width= barWidth, label='FB')
plt.bar(xpos, yt, bottom=fb, color='red', width= barWidth, label='YT')
plt.bar(xpos, ig, bottom=yt, color='purple', width= barWidth, label='IG')
plt.xlabel('Day')
plt.ylabel('Time spend in minutes')
plt.title('Time spend on socia media')
plt.xticks(xpos, x)
plt.legend()
plt.show()
```
